```python
x=[1, 2, 3, 4, 5, 6, 7, 8]
y=[32, 38, 45, 52, 60, 68, 75, 82]
```

```python
class linear_regression:
    def __init__(self):
        self.w=0
        self.b=0
        self.learning_rate=0.01

    def cost_function(self,x,y):
        t_sum=0;
        for i,j in zip(x,y) :
            y_pred= self.w*i +self.b
            t_sum+=(y_pred-j)**2
        t=(t_sum)/(2*len(x))
        return t
    def wgradient_function(self,x,y):
        grad_sum=0;
        for i,j in zip(x,y) :
            y_pred= self.w*i +self.b
            grad_sum+=i*(y_pred-j)
        grad=grad_sum/len(x)
        return grad
    def bgradient_function(self,x,y):
        grad_sum=0;
        for i,j in zip(x,y) :
            y_pred= self.w*i +self.b
            grad_sum+=(y_pred-j)
        grad=grad_sum/len(x)
        return grad
    def train(self,x,y,n):

        self.w=0
        self.b=0
        m=1
        for i in range(n):
            cost=cost_function(x,y,self.w,self.b)
            w_grad=self.wgradient_function(x,y)
            b_grad=self.bgradient_function(x,y)
            self.w+= -1*self.learning_rate*w_grad
            self.b += -1*self.learning_rate*b_grad
            print(f'{m} : loss = {cost} : gradient={w_grad,b_grad} : weight={self.w} : b={self.b}')
            m=m+1


    def predict(self,x):
        ans=[]
        for i in x:
            ans.extend([ i*self.w +self.b])
        return ans

    
        
```

```python
m=linear_regression()
```

```python
m.train(x,y,10000)
```

```python
m.predict([9,10,11,12])
```
